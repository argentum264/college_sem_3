# <center> One Hot Encoding

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Loading the iris dataset
iris = load_iris()

# Creating a DataFrame from the iris data
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Adding the target column as a categorical column
df['species'] = iris.target

df.head()
```

```python
# Map the numeric target values to species namee
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Final DF
print("Original DataFrame with Categorical Column:\n")
df.head()
```

```python
# Perform One-Hot Encoding on the species column
one_hot_encode = pd.get_dummies(df['species'], prefix='species')

# Dropping original species column and concatinating the final encoded columns
df_encoded = pd.concat([df.drop('species', axis=1), one_hot_encode], axis=1)

# Final DF
print("DataFrame after One-Hot Encoding:")
df_encoded.head()
```
