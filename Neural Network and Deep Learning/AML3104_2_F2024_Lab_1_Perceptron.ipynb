# Perceptron

```python
import random
```

```python
# OR Gate operations
input_data = [(0, 0), (0, 1), (1, 1)]
output_data = [0, 1, 1]
input_data, output_data
```

```python
# Initialize weight
w1 = random.uniform(-1, 1)
w2 = random.uniform(-1, 1)
b = random.uniform(-1, 1)
w1, w2, b
```

```python
# Learning rate
lr = 0.01
# Nuber of trials
epochs = 100
```

```python
# Define activation function
def activate(weighted_sum):
  threshold = 0
  if weighted_sum >= threshold:
    return 1
  else:
    return 0
```

```python
for epoch in range(epochs):
  total_error = 0
  print("Epoch: ", epoch)

  for i in range(len(input_data)):
    x1, x2 = input_data[i]
    y = output_data[i]
    print("x1: ", x1, "x2: ", x2, "Expected output: ", y)

    # Compute weighted sum
    weighted_sum = x1*w1 + x2*w2 + b
    print("Weighted sum: ", weighted_sum)
    # Compute activation
    output = activate(weighted_sum)
    print("Output: ", output)
    # Compute error
    error = y - output
    print("Error: ", error)

    # Total error
    total_error += abs(error)
    print("Total error: ", total_error)

    # Update the wights and bias
    w1 += lr*error*x1
    w2 += lr*error*x2
    b += lr*error
    print("Updated weights: ", w1, w2, b)
    print()

  if total_error == 0:
    print("\n Model learned at epoch: ", epoch)
    break



```

```python
# Learned Weights
print("Learned weights: ", w1, w2, b)
```

```python
# OR gate test data
test_data = [(0, 0), (0, 1), (1, 0), (1, 1)]
expected_test_output = [0, 1, 1, 1]
test_data, expected_test_output
```

```python
for i in range(len(test_data)):
  x1, x2 = test_data[i]
  y = expected_test_output[i]
  weighted_sum = x1*w1 + x2*w2 + b
  output = activate(weighted_sum)
  print(f"Input: ({x1}, {x2}) - Predicted_output: {output}, Expected_output: {y}")
```
